```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("/content/behaviour_simulation_train.xlsx - Sheet1 (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['datetime'] = pd.to_datetime(df['date'], errors='coerce') #creates newcolumn of datetime so that no afffect on date original column
#of changes we will do to seperate it into hour day
df['has_media'] = df['media'].astype(bool)# creates a new column to check if it has media
df['has_media'] = df['has_media'].astype(int)  # Column becomes [1, 0, 1, 0]

df['day_of_week'] = df['datetime'].dt.day_name()
df['hour'] = df['datetime'].dt.hour
df = pd.get_dummies(df, columns=['day_of_week',"inferred company"], drop_first=True)

df.head()
```

```python
df.head()
```

```python
median_like=df["likes"].median()
median_like
```

```python
df['likes_category'] = (df['likes'] >= median_like).astype(int)  # 1: High likes, 0: Low likes
df.head()
```

```python
df['inferred company_waitrose'] = df['inferred company_waitrose'].astype(int)
df['inferred company_walmart'] = df['inferred company_walmart'].astype(int)
df['inferred company_wells fargo'] = df['inferred company_wells fargo'].astype(int)
df['inferred company_western union'] = df['inferred company_western union'].astype(int)
df['inferred company_williams'] = df['inferred company_williams'].astype(int)
df['inferred company_wwf'] = df['inferred company_wwf'].astype(int)
df['inferred company_yahoo'] = df['inferred company_yahoo'].astype(int)
df['inferred company_ymca'] = df['inferred company_ymca'].astype(int)
df['inferred company_zumba'] = df['inferred company_zumba'].astype(int)
df['day_of_week_Monday'] = df['day_of_week_Monday'].astype(int)
df['day_of_week_Tuesday'] = df['day_of_week_Tuesday'].astype(int)
df['day_of_week_Wednesday'] = df['day_of_week_Wednesday'].astype(int)


df['day_of_week_Thursday'] = df['day_of_week_Thursday'].astype(int)
df['day_of_week_Saturday'] = df['day_of_week_Saturday'].astype(int)
df['day_of_week_Sunday'] = df['day_of_week_Sunday'].astype(int)
df.head()
```

```python
from textblob import TextBlob

df['sentiment'] = df['content'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['char_count'] = df['content'].apply(len)
df["char_count"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
model = LogisticRegression()

x=df.drop(columns=["date","content","username","media","day_of_week","likes","datetime"])
y=df["likes_category"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.head()
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
y_pred_new=rf.predict(x_test)
accuracy=accuracy_score(y_test,y_pred_new)
print(accuracy)
```
